### 1. Find the average age of developers when they wrote their first line of code. ###

```python
data['Age1stCode'] = pd.to_numeric(data['Age1stCode'], errors='coerce')
avg_age_first_code = data['Age1stCode'].astype(float).mean()
print("Average age of developers when they wrote their first line of code:", avg_age_first_code)
```

### 2. Deduce the percentage of developers who know python in each country.

```python
python_users_by_country = data[data['LanguageWorkedWith'].str.contains('Python', na=False)]

percentage_of_python_users_by_country = (python_users_by_country['Country'].value_counts() / data['Country'].value_counts()) * 100

print("Percentage of developers who know Python in each country:")
print()
print(percentage_of_python_users_by_country)
```

### 3. In which countries do developers work the highest number of hours per week? Consider countries with more than 250 responses only.

```python
countries_with_250_responses = data['Country'].value_counts()[data['Country'].value_counts() > 250].index

highest_hours_by_country = data[data['Country'].isin(countries_with_250_responses)] \
    .groupby('Country')['WorkWeekHrs'] \
    .mean() \
    .sort_values(ascending=False)

print("Countries where developers work the highest number of hours per week:")
print()
print(highest_hours_by_country)
```

### 4. Based on this survey, what will be the most desired programming language for the year 2020?

```python
desired_languages_2020 = data['LanguageDesireNextYear'].str.split(';', expand=True).stack()
desired_languages_counts = desired_languages_2020.value_counts()
most_desired_language_2020 = desired_languages_counts.idxmax()
print("The most desired programming language for the year 2020:", most_desired_language_2020)
```

### 5. What is the distribution of people who code as a hobby based on gender and continent (hint: use your mapping of country to continent)?

```python
hobbyist_distribution = data.groupby(['Country', 'Gender'])['Hobbyist'] \
    .value_counts(normalize=True) * 100

print("Distribution of people who code as a hobby based on gender and country:")
print()
print(hobbyist_distribution)
```

### 6. Generate the report for job and career satisfaction of developer based on their gender and continent?

```python
satisfaction_report = data.groupby(['Country', 'Gender', 'JobSat', 'CareerSat']).size() \
    .unstack(fill_value=0) \
    .apply(lambda x: x / x.sum() * 100, axis=1)

print("Job and career satisfaction of developers based on gender and country:")
print()
print(satisfaction_report)
```
